```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Employee.csv')
```

```python
df_reset=df.reset_index(drop=True)
df_reset.index=df_reset.index+1
df_reset.head()
```

```python
sns.countplot(x="City",data=df_reset)
```

```python
df["Education"].value_counts().plot(kind="pie",autopct="%.2f")
```

```python
sns.distplot(x=df["Age"])
```

```python
scatterplot=sns.scatterplot(x="Age",y="ExperienceInCurrentDomain",data=df_reset)
```

```python
p=pd.crosstab(df_reset["City"],df_reset["PaymentTier"])
print(p)
```

```python
sns.heatmap(p)
```

```python
sns.histplot(x="Age",hue="City",data=df_reset)
```

```python
fig,axes = plt.subplots(1,2,figsize=(12, 6))
sns.histplot(data=df_reset, x="ExperienceInCurrentDomain", bins=15, kde=True, ax=axes[1], color="green")
```

```python
leave_rate = df_reset.groupby("JoiningYear")["LeaveOrNot"].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=leave_rate, x="JoiningYear", y="LeaveOrNot", marker="o", color="purple")
plt.title("Line Chart: Average Leave Rate Over Years")
plt.ylabel("Average Leave Rate")
plt.xlabel("Joining Year")
plt.grid(True)
plt.show()
```

```python
numeric_data = df_reset.select_dtypes(include=["int64", "float64"])
corr_matrix = numeric_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Heatmap: Correlation Matrix")
plt.show()
```
