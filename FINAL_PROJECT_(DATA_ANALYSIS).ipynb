```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Player': ['Virat Kohli', 'Rohit Sharma', 'Steve Smith', 'Kane Williamson'],
    'Country': ['India', 'India', 'Australia', 'New Zealand'],
    'Runs': [12000, 10000, 9500, 8000],
    'Average': [50.0, 45.0, 55.0, 50.0]
}

df = pd.DataFrame(data)

sns.set(style="darkgrid")

plt.figure(figsize=(8,5))
sns.barplot(x='Player', y='Runs', data=df, palette='coolwarm')
plt.title('INTERNATIONAL RUNS', fontsize=16)
plt.xlabel('PLAYER', fontsize=14)
plt.ylabel('Runs', fontsize=14)
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x='Player', y='Average', data=df, palette='magma')
plt.title('BATTING AVERAGE', fontsize=16)
plt.xlabel('PLAYER', fontsize=14)
plt.ylabel('Average', fontsize=14)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extended Data with region-wise scores
data = {
    'Player': ['Virat Kohli', 'Rohit Sharma', 'Steve Smith', 'Kane Williamson'],
    'Country': ['India', 'India', 'Australia', 'New Zealand'],
    'Asia': [6000, 5000, 3000, 2000],
    'Australia': [2500, 2000, 4500, 3000],
    'Europe': [1500, 1000, 1000, 1500],
    'Africa': [2000, 2000, 1000, 1500],
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Melting the DataFrame to a long format for better visualization with seaborn
df_melt = pd.melt(df, id_vars=['Player', 'Country'], value_vars=['Asia', 'Australia', 'Europe', 'Africa'],
                  var_name='Region', value_name='Runs')

# Set the style
sns.set_style('whitegrid')

# Plotting region-wise scores using a barplot
plt.figure(figsize=(12,6))
sns.barplot(x='Player', y='Runs', hue='Region', data=df_melt, palette='Spectral')
plt.title('Region-Wise Scores by Players', fontsize=16)
plt.xlabel('Player', fontsize=14)
plt.ylabel('Runs', fontsize=14)
plt.legend(title='Region')
plt.show()
```
